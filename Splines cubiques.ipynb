```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import CubicSpline as spline
```

```python
N = 100
X = [0,1,2,3]
U = [10,11,2,15]
x = np.linspace(X[0],X[-1], N)

Y_spline = spline(X,U)(x)

Y_line = np.empty_like(Y_spline)
k = 0
Y_line[0] = U[0]
for i in range(len(U)-1):
    space = 100/(len(U)-1)
    print(space)
    inc = (U[i+1] - U[i])/(space)
    for j in range(int(space)):
        Y_line[k+1] = Y_line[k] + inc
        k+= 1
        
        
    
```

```python
ax = plt.subplot(1, 1, 1)
ax.plot(x, Y_spline, label='Spline cubique')
ax.plot(x, Y_line, label='Spline linéaire')

ax.scatter(X,U)
ax.legend()
#plt.ylim([-0.25,2])
plt.show()
```
